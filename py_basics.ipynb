### 3.1.1 Condicional if

```python
## programa que dice el numero mayor
numero1 = int(input("Dime un numero..."))
numero2 = int(input("Dime otro numero..."))

# si numero1 es mayor que numero2 entonces imprime numero1 es mayor que numero2  sino imprime numero1 es menor que numero2
if numero1 > numero2:
    print(f"{numero1} es mayor que {numero2}")
elif numero1 < numero2:
    print(f"{numero1} es menor que {numero2}")
```

```python
## programa que dice si un numero es par o impar
numero = int(input("Dime un numero..."))

#  si el residuo de numero entre 2 es igual a 0 entonces imprime numero es par sino imprime numero es impar
if numero % 2 == 0:
    print(f"{numero} es par")
else:
    print(f"{numero} es impar")
```

```python
## programa que dice si un numero es positivo o negativo
numero = int(input("Dime un numero..."))

# si numero es mayor que 0 entonces imprime numero es positivo sino si numero es menor que 0 entonces imprime numero es negativo sino imprime numero es cero
if numero > 0:
    print(f"{numero} es positivo")
elif numero < 0:
    print(f"{numero} es negativo")
else:
    print(f"{numero} es cero")
```

```python
# programa que dice si una persona es mayor o menor de edad
edad = int(input("Dime tu edad..."))

# si edad es mayor o igual que 18 entonces imprime eres mayor de edad sino imprime eres menor de edad
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")

```

### 3.2 Bucle while

```python
#imprimir los numeros del 1 al 5

contador = 1
while contador <= 5:
    print(contador)
    contador = contador + 1
    
```

```python
#imprimir los numeros del 5 al 1
contador = 5
while contador >= 1:
    print(contador)
    contador = contador - 1
```

```python
# programa que sale del bucle hasta que se adivina un numero

"""se crea una variable numero con el valor 10, se crea un bucle infinito y se pide al usuario que adivine el numero, si el numero que adivino el usuario es igual al numero entonces imprime adivinaste el numero y rompe el bucle sino imprime no adivinaste el numero
"""
numero = 10
while True:
    numero_adivinar = int(input("Adivina el numero..."))
    if numero_adivinar == numero:
        print("Adivinaste el numero")
        break
    else:
        print("No adivinaste el numero")
```

```python
#programa que saca el promedio de notas de un alumno

"""se crea una variable contador con el valor 0, se crea una variable suma con el valor 0, se crea un bucle infinito y se pide al usuario que ingrese una nota, se suma la nota a la variable suma, se aumenta en 1 la variable contador, se pregunta al usuario si desea ingresar otra nota, si la respuesta es no entonces se imprime el promedio de notas y se rompe el bucle
"""
contador = 0
suma = 0
while True:
    nota = float(input("Ingrese una nota..."))
    suma = suma + nota
    contador = contador + 1
    respuesta = input("Desea ingresar otra nota? (si/no)")
    if respuesta == "no":
        print(f"El promedio de notas es {suma/contador}")
        break

```

```python
numeros = [10, 5, 7, 2, 1]
print("Contenido de la lista:", numeros)  # Imprimiendo contenido de la lista.
 
# agregar un elemento al final de la lista
numeros[0] = 111
print("Nuevo contenido de la lista: ", numeros)  # Contenido actual de la lista.

# agregar el valor del quinto elemento al segundo elemento
numeros[1] = numeros[4]  # Copiando el valor del quinto elemento al segundo elemento.
print("Nuevo contenido de la lista:", numeros)  # Imprimiendo el contenido de la lista actual.

print("\nLongitud de la lista:", len(numeros))  # Imprimiendo la longitud de la lista.
```

```python
# acceder a los elementos usando su indice

lista = [17, 3, 11, 5, 1, 9, 7, 25, 13]
print(lista[0])
print(lista[1])

# usando rango del indice
lista = [17, 3, 11, 5, 1, 9, 7, 25, 13]
print(lista[0:3])
print(lista[2:5])
print(lista[5:8])
```

```python
# encontrar el numero mas largo de una lista
lista = [17, 3, 11, 5, 1, 9, 7, 25, 13]
numero_largo = lista[0]

for i in range(1, len(lista)):
    if lista[i] > numero_largo:
        numero_largo = lista[i]

print(numero_largo)
```

```python
# encontrar el numero mas corto de una lista

lista = [17, 3, 11, 5, 1, 9, 7, 25, 13]
numero_corto = lista[0]

for i in range(1, len(lista)):
    if lista[i] < numero_corto:
        numero_corto = lista[i]

print(numero_corto)
```

### 4.3 Diccionarios

```python
#diccionario con los personajes de stardew valley
personajes = { "Sebastian": "Programador", "Sam": "Músico", "Abigail": "Jugadora de rol", "Alex": "Atleta", "Haley": "Modelo"} 

print("------------------")
#imprimir el valor de un personaje
print("Profesión de Sam:")
print(personajes["Sam"])

print("------------------")
#agregar un nuevo personaje
personajes["Elliott"] = "Escritor"
print("diccionario con el personaje agregado:")
print(personajes)

print("------------------")
# agregar un nuevo personaje con update
personajes.update({"Harvey": "Doctor"})
print("Diccionario con el personaje agregado:")
print(personajes)

print("------------------")
#eliminar un personaje
del personajes["Alex"]
print("Diccionario con el personaje eliminado:")
print(personajes)
print("------------------")
```

```python
#diccionario con los personajes de stardew valley
personajes = { "Sebastian": "Programador", 
              "Sam": "Músico",
              "Abigail": "Jugadora de rol",
              "Alex": "Atleta",
              "Haley": "Modelo"} 


print("-------------------------------------")
print("Con un orden aburrido:")
print("-------------------------------------")

# acceder a los elementos de un diccionario
for key in personajes.keys():
  print(key, "--", personajes[key])

print("----------------------------------------")
print("Con mejor orden:")
print("----------------------------------------")


# acceder a los elementos de un diccionario decorado(?)
for key in personajes.keys():
  print("|", key, "--", personajes[key], "|")
```

```python
# acceder a los elementos de un diccionario con items
personajes = { "Sebastian": "Programador", 
              "Sam": "Músico",
              "Abigail": "Jugadora de rol",
              "Alex": "Atleta",
              "Haley": "Modelo"} 

print("--------------------")
for key, value in personajes.items():
  print(key, ":")
  print(value)
  print("--------------------")


```

```python
personajes = { "Sebastian": "Programador", 
              "Sam": "Músico",
              "Abigail": "Jugadora de rol",
              "Alex": "Atleta",
              "Haley": "Modelo"} 

# ordenar  un diccionario con sorted
for key in sorted(personajes.keys()):
  print(key, "--", personajes[key])

  
```
